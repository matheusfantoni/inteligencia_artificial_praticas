```python
import pandas as pd
import numpy as np
```

```python
application = pd.read_csv("application_record.csv", sep=',')
credit = pd.read_csv("credit_record.csv", sep=',')
```

1 - Join entre os arquivos da base de dados e salvando o resultado em um novo arquivo csv.

```python
resultado_csv = pd.merge(application, credit, on='ID', how='inner')

resultado_csv.to_csv('file.csv')
```

```python
resultado_csv.head()
```

2 - Criação de novas variáveis a partir das variáveis presentes na base de dados.

Campos Criados: YEAR_AGE / YEAR_EMPLOYED / AMT_INCOME_MONTH

```python
resultado_csv = resultado_csv.assign(YEAR_AGE=0)
print(resultado_csv)
```

```python
resultado_csv['YEAR_AGE'] = resultado_csv['DAYS_BIRTH'].apply(lambda x:x/-365)
resultado_csv.head()
```

```python
resultado_csv = resultado_csv.assign(YEAR_EMPLOYED=0)
resultado_csv.head()
```

```python
resultado_csv['YEAR_EMPLOYED'] = resultado_csv['DAYS_EMPLOYED'].apply(lambda x:x/-365)
resultado_csv.head()
```

```python
resultado_csv = resultado_csv.assign(AMT_INCOME_MONTH=0)
resultado_csv.head()
```

```python
resultado_csv['AMT_INCOME_MONTH'] = resultado_csv['AMT_INCOME_TOTAL'].apply(lambda x:x/12)
resultado_csv.head()
```
