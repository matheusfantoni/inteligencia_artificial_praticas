1° Gráfico - Contagem por Gênero Masculino e Feminino

3° Gráfico - Quantidade de Clientes que possuem dívidas vencidas ou incobráveis, baixas por mais de 150 dias.

Esta análise visa passar informações de como o status das dívidas dos clientes estão de acordo com a coluna STATUS do banco de dados da credit_record. As informações são analisadas da seguinte maneira: 

5: dívidas vencidas ou incobráveis, baixas por mais de 150 dias 


Concluímos que temos 1693 pessoas que possuem dívidas vencidas ou incobráveis.

```python
dataset2 = pd.read_csv("credit_record.csv", sep = ',')

dataset2.loc[dataset2['STATUS'] == '5'].value_counts("STATUS")
```

```python
dataset2.loc[dataset2['STATUS'] == '5'].value_counts("STATUS").plot.bar()
```

6° Gráfico - Contagem de pessoas que não possuem nenhum emprestimo 

Esta analise visa a obter informações de quantas pessoas que não possuem nenhum emprestimo para uma futura aprovação de um cartão de crédito solicitado. Essa informação é importante porque quanto mais pessoas não possuirem um emprestimo, menos chances poderam de ter dívidas inadimplentes. Usando a base de dados credit_record e o campo STATUS observaremos o número de pessoas contando os valores, conforme descrito abaixo são 209.230 clientes que não possuem nenhum emprestimo para este mês.

```python
dataset2.loc[dataset2['STATUS'] == 'X'].value_counts("STATUS")
```

```python
dataset2.loc[dataset2['STATUS'] == 'X'].value_counts("STATUS").plot.bar(color = 'red', title="Contagem de pessoas que não possuem nenhum emprestimo")
```
