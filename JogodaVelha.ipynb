```python
import random


def jogar():
    imprime_msg_abertura()

    start = sorteador_de_inicio()

    demonstracao_de_posicao()

    matriz = [["_","_","_"],
              ["_","_","_"],
              ["_","_","_"]]

    ordem = primeira_jogada(start, matriz)

    while (not fim_jogo(matriz)):
        if(ordem == 0):
            print("*** Sua vez! ***")
            print("")
            linha, coluna = get_input(matriz)
            matriz[linha][coluna] = "X"
            mostra_matriz(matriz)
            print("")
            ordem = 1
        else:
            print("\n*** Vez da CPU ***")
            linha, coluna = random.choice(get_lista_posicoes_possiveis(matriz))
            matriz[linha][coluna] = "O"
            mostra_matriz(matriz)
            print("")
            ordem = 0



def imprime_msg_abertura():
    print("*********************************")
    print("Iniciando o Jogo Da Velha!")             ###### Somente mensagens
    print("********************************* \n")


def sorteador_de_inicio():
    on = int(input("Aperte: (1) Iniciar o Jogo da Velha (2) Não\n"))
    
    while(on != 1):
        on = int(input("Já está pronto? Aperte 1 para iniciar o jogo: "))
    
    print("Realizando sorteio para definir se você ou a CPU irá começar a jogar.")
    
    sorteio = random.randrange(0,2)
    
    if(sorteio == 1):
        print("Você vai começar jogando e usara o X")
        print("CPU usuará o O")
        print("")
    else:
        print("A CPU começa usando O")
        print("Você pode ser identificado pelo X")
        print("")
    return sorteio


def fim_jogo(matriz):
    if verifica_vitoria(matriz):
        print("Ganhou")
        return True
    if verifica_derrota(matriz):
        print("Perdeu")
        return True
    if verifica_velha(matriz):
        print("Velha")
        return True
    return False


def verifica_vitoria(matriz):    ######## Implementar
    return False


def verifica_derrota(matriz):          ##### Implementar
    return False


def verifica_velha(matriz):
    if len(get_lista_posicoes_possiveis(matriz)) == 0:
        return True
    return False


def get_input(matriz):
    posicoes_disponiveis = get_lista_posicoes_possiveis(matriz)
    while True:
        try:
            linha = int(input("Insira o numero da linha: "))
            coluna = int(input("Insira o numero da coluna: "))
            if (linha, coluna) in posicoes_disponiveis:
                return (linha, coluna)
            print("Posição já marcada ou invalida")
        except ValueError:
            print("Erro")


def mostra_matriz(matriz):
    for n in range(0, 3):
        print(matriz[n])


def demonstracao_de_posicao():
    print("Essa é a demonstração das posições possiveis:\n")
    print("0,0","|","0,1","|","0,2")
    print("1,0","|","1,1","|","1,2")
    print("2,0","|","2,1","|","2,2")
    print("")


def primeira_jogada(start, matriz):
    if(start == 1):
        print("*** Vez do Jogador ***")
        linha, coluna = get_input(matriz)
        matriz[linha][coluna] = "X"
        mostra_matriz(matriz)
        print("")
        ordem = 1
    else:
        print("*** Vez da CPU ***")
        linha, coluna = random.choice(get_lista_posicoes_possiveis(matriz))
        matriz[linha][coluna] = "O"
        mostra_matriz(matriz)
        print("")
        ordem = 0
    return ordem


def get_lista_posicoes_possiveis(matriz):
    result = []
    for i in range(3):
        for j in range(3):
            if matriz[i][j] == "_":
                result.append((i, j))
    return result


if (__name__ == "__main__"):
    jogar ()
```
